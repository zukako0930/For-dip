````python
def get_tweet(usid):
    tweet_list=[]
    # ã‚¿ã‚¤ãƒ ãƒ©ã‚¤ãƒ³å–å¾—ç”¨ã®URL
    url = "https://api.twitter.com/1.1/statuses/user_timeline.json?count=100"
    # ã¨ãã«ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¯ç„¡ã„
    params = { "user_id" : usid}
    # OAuth ã§ GET
    twitter = OAuth1Session(CK, CS, AT, AS)
    req = twitter.get(url, params = params)
    n = 0
    # print(users["id_str"])
    if req.status_code == 200:
        # ãƒ¬ã‚¹ãƒãƒ³ã‚¹ã¯JSONå½¢å¼ãªã®ã§ parse ã™ã‚‹
        timeline = json.loads(req.text)
        # å„ãƒ„ã‚¤ãƒ¼ãƒˆã®æœ¬æ–‡ã‚’è¡¨ç¤º
        for tweet in timeline:
            tweet_list.append(tweet["text"])
        return tweet_list
#         print(tweet_list)        
    else:
        # ã‚¨ãƒ©ãƒ¼ã®å ´åˆ
        print ("Error: %d" % req.status_code)
        return 0
````

````python
def calc_score(tweet_list):
    score=0
    sum_score=0
    max_score=-1
    min_score=1
    for idx,tweet in enumerate(tweet_list):
#         print(idx)
        tw1=nega_posi.JudgeClass(tweet)
        tw1.make_wlist()
        score_dict=tw1.make_dict()
        tw1.make_judge_list()
        score=tw1.calc_score(score_dict)
        if score >max_score:
            max_idx=idx
            max_score=score
        if score < min_score:
            min_idx=idx
            min_score=score
        sum_score+=score
    if len(tweet_list) == 0:
        return 0
    else:
        print("å…¨ä½“å¹³å‡"+str(sum_score/len(tweet_list)))
        #scorelistã‹ã‚‰ä¸Šä½ã®indexã‚’å–å¾—
        print("positive:"+ tweet_list[max_idx])
        print("negative:"+tweet_list[min_idx])
        return sum_score/len(tweet_list)
````
