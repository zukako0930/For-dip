```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

from requests_oauthlib import OAuth1Session
from collections import defaultdict
import json

CK = '3rJOl1ODzm9yZy63FACdg'                             # Consumer Key
CS = '5jPoQ5kQvMJFDYRNE8bQ4rHuds4xJqhvgNJM4awaE8'         # Consumer Secret
AT = '2956677620-D7KuhxlChnM27XgQ9WoFJR4haZuTPgTnBVrKe1K' # Access Token
AS = '2qTkh1PC1OIL6rCPyNKUnmHV8Rf70XCODczCTFdp5n1uq'         # Accesss Token Secert

def get_follower_name(screen_name = "12satouT"):
    f_name = []
    cursor_init = -1
    
    while cursor_init != 0:
        url = "https://api.twitter.com/1.1/followers/list.json"
    

        params = { "screen_name" : screen_name,
                            "count": 200,
                            "cursor":cursor_init,
                         }

        twitter = OAuth1Session(CK, CS, AT, AS)
        req = twitter.get(url, params = params)

        n = 0

        if req.status_code == 200:
            timeline = json.loads(req.text)
            cursor_init = timeline["next_cursor"]
            for follower in timeline["users"]:
                f_name.append(follower["screen_name"])
        else:
            print ("Error: %d" % req.status_code)
        
        cursor_init = 0
    
    file = [line+"\n" for line in f_name]
    with open('./data/follower/{}.txt'.format(screen_name), 'w', encoding='utf8') as f:
        f.writelines(file)
    
    return f_name

```
