### count_listの平均と標準偏差を取る

```python
import numpy as np
import math
def calc_std(count_list):
    np_list=np.array(count_list)
    sc=np_list-ave
    sc2=np.power(sc,2)
    sc3=np.sum(sc2)

    std=math.sqrt(sc3/len(np_list))
    return std
```

```python
ave=sum(count_list)/len(count_list)
print(ave)
std = calc_std(count_list)
print(std)
```
