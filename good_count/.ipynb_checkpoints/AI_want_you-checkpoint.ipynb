### スクリーンネームからいいね数を取ってくる

```python
def home_line(screen_name):
    url = "https://api.twitter.com/1.1/statuses/user_timeline.json?count=100"
#     url = "https://api.twitter.com/1.1/statuses/mentions_timeline.json"
    favolist = []
    
    params = {
        "screen_name": screen_name,
#         "count": ,
        "include_entities": True,
    }
    req = twitter.get(url, params=params)
    result_2 = json.loads(req.text)
    for repcount in result_2:
#         print(repcount)
        if  '@' not in repcount['text']:
        
            favolist.append(repcount['favorite_count'])
    # uni_tweet = tweet.decode('utf-8')
    return favolist
```

### 実際の計算

```python
ave_list=[]
for u in scnames:
    try:
        print(u)
        favolist = home_line(u)
        ave = sum(favolist)/len(favolist)
        ave_list.append(ave)
    except:
        continue
print(ave_list)
```

### 一般人なので平均いいねが100以下の人を選ぶ

```python
new_ave_list=[]
new_ave_list=[x for x in ave_list if x < 100]
print(new_ave_list)
print(len(new_ave_list)) # 1725人
```

### 平均

```python
print(sum(new_ave_list)/len(new_ave_list))
ave=sum(new_ave_list)/len(new_ave_list)
```

### 分散

```python
import numpy as np
npave_list=np.array(new_ave_list)
sc=npave_list-ave
sc2=np.power(sc,2)
sc3=np.sum(sc2)

import math
std=math.sqrt(sc3/len(npave_list))
std
```
