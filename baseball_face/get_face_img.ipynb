```python
#coding:utf-8 
from bs4 import BeautifulSoup #bs4モジュールからBeautifulSoupクラスをインポートする
import requests
import csv
import scrapelib
import os
from urllib import request
```

```python
# url先の画像を保存する関数
def download(url):
    img = request.urlopen(url)
    localfile = open(os.path.basename(url), 'wb')
    localfile.write(img.read())
    img.close()
    localfile.close()
```

```python
s = scrapelib.Scraper(requests_per_minute=30) #１分間に何回リクエストを投げるか
t=s.get('http://npb.jp/bis/teams/rst_d.html')
soup=BeautifulSoup(t.text,'lxml')
```

```python
ss=soup.select('.rosterPlayer')
```

```python
#選手のURLの取得
link_list=[]
for a in ss:
    href = a.find('a')
    if(href == None):
        continue
    else:
        link_list.append(href.attrs['href'])
print(link_list)
```

```python
for link in link_list:
    print(link)
    t2=s.get('http://npb.jp/'+str(link))
    soup2=BeautifulSoup(t2.text,'lxml')    
    ss2=soup2.select('.rgshdphoto')
    img_list=[]
    for img in ss2:
        imglink = img.find('img')
        if(href == None):
            continue
        else:
            img_list.append(imglink.attrs['src'])
    print(img_list)

    download(img_list[0])
```
